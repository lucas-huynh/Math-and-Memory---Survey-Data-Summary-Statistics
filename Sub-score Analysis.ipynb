```python
p_jobs = 'Jobs/Careers (parent)'
df = pd.read_csv(file_path)
p_jobs_values = df[p_jobs].tolist()

p_worse_job = 0
p_same_job = 0
p_better_job = 0

p_jobs_val_list = []

for val in p_jobs_values:
    if val == 'Less/worse quality':
        p_worse_job += 1
        p_jobs_val_list.append(1/15)
    elif val == 'The same amount/quality':  
        p_same_job += 1
        p_jobs_val_list.append(2/15)
    elif val == 'More/better quality':
        p_better_job += 1
        p_jobs_val_list.append(3/15)

print("Worse Jobs (Parent) count:", p_worse_job)
print("Same Jobs (Parent) count:", p_same_job)
print("Better Jobs (Parent) count:", p_better_job)


p_jobs_array = np.array(p_jobs_val_list)
p_jobs_count = len(p_jobs_array)
p_jobs_mean = np.mean(p_jobs_array)
p_jobs_std = np.std(p_jobs_array)
p_jobs_med = np.median(p_jobs_array)
p_jobs_min = np.min(p_jobs_array)
p_jobs_max = np.max(p_jobs_array)

print('')
print('Job Quality Summary Statistics (Parent)')
print('n =', p_jobs_count)
print('Mean =', p_jobs_mean)
print('Std_Dev =', p_jobs_std)
print('Median =', p_jobs_med)
print('Min =', p_jobs_min)
print('Max =', p_jobs_max)
```
