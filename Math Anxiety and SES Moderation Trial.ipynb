```python
from pyprocessmacro import Process
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
df = pd.read_csv("SES_MA.csv")
p = Process(data=df, model=1, x= "Standardized Ladder (child)", y= "Math Anxiety (Child)", m= "Standardized Ladder (parent)")
p.summary()
```

# Variable Definitions
- X --> Standardized MacArthur Social Ladder Score for Child (scale of 0.0 to 1.0)
- Y --> Child Math Anxiety Score (scale 1 to 10)
- M --> Standardized MacArthur Social Ladder Score for Child (scale of 0.0 to 1.0)

# Formula
Y = B0 + B1X + B2M + B3XM
- MA = 0.6389 - 1.0604(Child Ladder) - 0.8481(Parent Ladder) + 2.3949(Child Ladder * Parent Ladder)

# Overall Model Fit
- R^2 = 0.1147
- Adjusted R^2 = 0.0004
- Interpretation: The model explains a small proportion of the variance

- MSE = 0.0323
- p = 0.2661

# Coefficients
- Child Ladder --> p = 0.0838 does not suggest a statistically significant effect of CHild Socal Ladder Scores on Child MA
- Parent Ladder --> p = 0.2366, therefore Parent Social Ladder does not have a statistically significant (given a = 0.05) effect on Child MA
- Child Ladder * Parent Ladder --> p = 0.1507, therefore the interaction terms does not have a statistically significant (given a = 0.05) effect on Child MA
